### STEP 1: DATA CURATION
First, let's extract the CSV data.<br>

Let's write a function called read_data that takes the data file path as an argument, uses pandas to read the csv file and outputs the first five entries using head(). The output is our data set in the form of a table with various columns as explained above.

#### Missing values and NaN
As we can tell from the above table, there are several missing values. This indicates that the data is either unavailable or has not been collected. These values are treated as noise in the data. For preliminary analysis, I chose to replace all the '?' with NaN and drop all the rows with such values. For extensive analysis, you can adopt imputation of missing values. ### STEP 4: HYPOTHESIS TESTING AND MODEL ANALYSIS
#### Hypothesis<br>
From our priliminary analysis we do not have sufficient evidence to conclude that the number of crashes is related to the year. I believe there might be some correlation between fatality count and one of the following: aircraft_type, route, and/or month. 
Which will be validated using a linear regression model and plotting residual plot. For linear regression, residual errors follow a normal distribution.
Let us look at the number of unique values for each of these. Since the unique counts are very large for aircraft_type and route we need to use hash encoding to avoid overfitting. A useful source on encoding data for categorical variables: https://towardsdatascience.com/smarter-ways-to-encode-categorical-data-for-machine-learning-part-1-of-3-6dca2f71b159 <br>
The number of unique values of month is 12, which is relatively less. For this, we can use one hot encoding.
<br>More on one hot encoding: https://hackernoon.com/what-is-one-hot-encoding-why-and-when-do-you-have-to-use-it-e3c6186d008f

```python
def one_hot_encoding(data=None, columns=None):
    '''
    one hot encoding for categorical data
    '''
    if data is None or columns is None:
        raise ValueError('Need Data and Columns as arguments !')
    data = data.copy()
    data = pd.concat([data, pd.get_dummies(data=data[columns], columns=columns)], axis=1)
    data.drop(columns, axis=1, inplace=True)
    return data
```

```python
def hash_encoding(data=None, columns=None, n_features=5, input_type='string'):
    '''
     Hash Encoding for categorical data
    '''
    if data is None or columns is None or n_features is None or input_type is None:
        raise ValueError(
            'Need Data, Columns, n_features and  input_type as arguments !')
    data = data.copy()
    hasher = FeatureHasher(n_features, input_type)
    label_1 = [columns[0] + '_' + str(j) for j in range(0, n_features)]
    label_2 = [columns[1] + '_' + str(j) for j in range(0, n_features)]
    hashed = hasher.fit_transform(data[columns[0]])
    df = pd.DataFrame(hashed.toarray(), columns=label_1)
    data = pd.concat([data, df], axis=1)
    data.drop(columns[0], axis=1, inplace=True)
    hashed = hasher.fit_transform(data[columns[1]])
    df = pd.DataFrame(hashed.toarray(), columns=label_2)
    data = pd.concat([data, df], axis=1)
    data.drop(columns[1], axis=1, inplace=True)
    return data
```

The following code divides the data into 70% and 30%. 70% for training data and 30% for testing data. The training set is implemented to build up a model, while a test set is to validate the model that has been built. Data points in the training set are excluded from the test set. A dataset is divided into a testing set and a training set in each iteration.

```python
def split_train_test_data(data=None, y_label=None, test_size=None, random_state=15):
    '''
    spilts data between training and validation and testing
    '''
    if data is None or y_label is None or test_size is None:
        raise ValueError('Need Data, Config map, Test size as arguments !')
    data = data.copy()
    y = data[y_label]
    X = data.drop(y_label, axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, y_train, X_test, y_test
```

The following code calculates the root mean square error. Root Mean Square Error = (y-y_pred)^2 
The root-mean-square error is a frequently used measure the differences between values ( population or a sample values) predicted by a model or an estimator and the values observed.

```python
def calculate_error(X=None, y=None, model=None):
    '''
    returns residual and rmse error for already fit model
    '''
    if X is None or y is None or model is None:
        raise ValueError('Need X, Y, Model as arguments !')
    predictions = model.predict(X)
    mse = mean_squared_error(y, predictions)
    rmse = np.sqrt(mse)
    return rmse
```

```python
def residual_plot(X_train=None, y_train=None, X_test=None, y_test=None):
    if X_train is None or y_train is None or X_test is None or y_test is None:
        raise ValueError('Need X_train, y_train, X_test, y_test as arguments !')
    model = LinearRegression(n_jobs=-1)
    visualizer = ResidualsPlot(model)
    visualizer.fit(X_train.values, y_train.values)
    visualizer.score(X_test.values, y_test.values)
    visualizer.poof()
    return None
```

```python
model_data = model_data[['month', 'route', 'aircraft_type', 'fatalities_count']]
model_data = one_hot_encoding(model_data, 'month')
model_data = hash_encoding(model_data, ["route", "aircraft_type"])
X_train, y_train, X_test, y_test = split_train_test_data(model_data, 'fatalities_count', 0.3)
```

The training error and testing error are converging to a value which implies the model has no bias (underfitting) and no variance (overfitting).

```python
#Model training
model = LinearRegression(n_jobs=-1)
model.fit(X_train, y_train)
training_rmse = calculate_error(X_train, y_train, model)
print ("training error")
print(training_rmse)
testing_rmse = calculate_error(X_test, y_test, model)
print ("testing error")
print(testing_rmse)
```

```python
#Residual Plot
residual_plot(X_train, y_train, X_test, y_test)
```

### STEP 5: CONCLUSION
From the residual plot we can notice that the residual errors follow a normal distribution as most of the data points are scattered around the axis. So our intial hypothesis of the features considered (month, route and aircraft_type) has some degree of correlation with fatality count.
